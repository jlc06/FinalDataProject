```python
import pandas as pd
def cleanffball(year,name):
    url = f'https://www.pro-football-reference.com/years/{year}/fantasy.htm'
    df = pd.read_html(url,header=1)[0]
    df = df.drop(['Rk','Tm','VBD'],axis=1)
    df = df[df['FantPos'].notna()]
    df = df[df['FantPos'] != 'FantPos']
    df['Y/A'] = df['Y/A'].fillna(0)
    df['Y/R'] = df['Y/R'].fillna(0)
    df['2PM'] = df['2PM'].fillna(0)
    df['2PP'] = df['2PP'].fillna(0)
    df = df[df['FantPt'].notna()]
    df = df.replace('\+','',regex=True).astype(object)
    df = df.replace('\*','',regex=True).astype(object)
    columns = df.columns[2:]
    for col in columns:
        df[col] = pd.to_numeric(df[col],errors='coerce')
    df['N_PPR_OvRank'] = df['FantPt'].rank(method='first',ascending=False)
    df['PPR_OvRank'] = df['PPR'].rank(method='first',ascending=False)
    df = df.drop(columns=['OvRank'],axis=1)
    df = df.dropna()
    df['Year'] = year
    df.to_csv(f'{name}.csv',index=False)
```

```python
years = [2021,2020,2019,2018,2017,2016,2015,2014,2013,2012]
for year in years:
    cleanffball(year,'season_' + str(year))
```

```python
season_21 = pd.read_csv('season_2021.csv')
season_20 = pd.read_csv('season_2020.csv')
season_19 = pd.read_csv('season_2019.csv')
season_18 = pd.read_csv('season_2018.csv')
season_17 = pd.read_csv('season_2017.csv')
season_16 = pd.read_csv('season_2016.csv')
season_15 = pd.read_csv('season_2015.csv')
season_14 = pd.read_csv('season_2014.csv')
season_13 = pd.read_csv('season_2013.csv')
season_12 = pd.read_csv('season_2012.csv')
seasons_L10Y = pd.concat([season_21,season_20,season_19,season_18,season_17,season_16,season_15,season_14,season_13,season_12], ignore_index=True)
seasons_L10Y
```
